### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 6)
y = x ** 2
bar_values = [5, 7, 3, 8, 6]
scatter_x = np.random.rand(10)
scatter_y = np.random.rand(10)
pie_labels = ['A', 'B', 'C', 'D']
pie_sizes = [30, 25, 25, 20]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart in the first subplot
axs[0, 0].plot(x, y, marker='o')
axs[0, 0].set_title("Line Chart")

# Bar chart in the second subplot
axs[0, 1].bar(x, bar_values, color='orange')
axs[0, 1].set_title("Bar Chart")

# Scatter plot in the third subplot
axs[1, 0].scatter(scatter_x, scatter_y, color='green')
axs[1, 0].set_title("Scatter Plot")

# Pie chart in the fourth subplot
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title("Pie Chart")
axs[1, 1].axis('equal')  # Equal aspect ratio for pie chart

plt.tight_layout()
plt.show()
```
